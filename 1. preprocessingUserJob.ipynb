```python
Tech_extract = [
    'LanguageWantToWorkWith',
    'DatabaseWantToWorkWith',
    'PlatformWantToWorkWith',
    'WebframeWantToWorkWith',
    'MiscTechWantToWorkWith',
    'ToolsTechWantToWorkWith',
    'NEWCollabToolsWantToWorkWith',
    'OpSysProfessional use',
    'OfficeStackAsyncWantToWorkWith',
    'OfficeStackSyncWantToWorkWith',
    'AISearchWantToWorkWith',
    'AIDevWantToWorkWith'
]
extracted_df = users_df[Tech_extract]
print(extracted_df.info())
print(extracted_df.head(3))
print(extracted_df.isna().sum())
```

```python
def concatenate_tech_columns(row):
    # Filter out None or NaN values before joining
    filtered_values = [str(value) for value in row if pd.notnull(value)]
    return ' '.join(filtered_values)

# Apply the function across the Tech_extract columns and create a new column for the merged data
extracted_df['Tech_Features_Merged'] = extracted_df[Tech_extract].apply(concatenate_tech_columns, axis=1)

# Drop all columns in Tech_extract
extracted_df.drop(columns=Tech_extract, inplace=True)
print(extracted_df.head(3))
```

```python
users_df = users_df.join(extracted_df)
users_df.info()
```

```python
users_df['Tech_Features_Merged'].replace('', np.nan, inplace=True)
missing_values = users_df['Tech_Features_Merged'].isnull().sum()

print("Missing values after treating empty strings as NaN:", missing_values)
```

```python
users_df.dropna(subset=['Tech_Features_Merged'], inplace=True)
users_df.reset_index(drop=True, inplace=True)
users_df.info()
```

```python
users_df['LearnCodeOnline'].isna().sum()
reduced_users_df = users_df.dropna(subset=['LearnCodeOnline'])
reduced_users_df.info()
```

```python
reduced_users_df = reduced_users_df.sample(n=50000, random_state=42)
reduced_users_df.to_csv('Data/reduced_users.csv', index=False)
```

```python
#doc2vec algo

!pip install gensim nltk
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python


def preprocess_text(text):
  text = text.replace(',', '').replace(';', '')
    # Tokenize text
  tokens = nltk.word_tokenize(text.lower())
  lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
  processed_text = ' '.join(lemmatized_tokens)
  return processed_text
```

```python
reduced_users_df['processed_text'] = reduced_users_df['Tech_Features_Merged'].apply(preprocess_text)
# reduced_jobs_df['processed_text'] = reduced_jobs_df['Job_Features_Merged'].apply(lambda x: ' '.join(preprocess_text(x)))
```
