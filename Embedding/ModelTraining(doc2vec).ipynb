```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import pandas as pd

all_texts = users_df['processed_text'].tolist() + jobs_df['processed_text'].tolist()
tagged_data = [TaggedDocument(words=text.split(), tags=[i]) for i, text in enumerate(all_texts)]
```

```python
model = Doc2Vec(vector_size=50, window=2, min_count=1, workers=4, epochs=40)
model.build_vocab(tagged_data)
model.train(tagged_data, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
# user_embeddings = [model.infer_vector(text.split()) for text in users_df['processed_text']]
# job_embeddings = [model.infer_vector(text.split()) for text in jobs_df['processed_text']]
users_df['embeddings'] = users_df['processed_text'].apply(lambda text: model.infer_vector(text.split()))
jobs_df['embeddings'] = jobs_df['processed_text'].apply(lambda text: model.infer_vector(text.split()))
```
