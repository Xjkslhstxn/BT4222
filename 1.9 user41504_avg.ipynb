```python
import pandas as pd
import numpy as np
```

```python
pred0_df = pd.read_csv('job_predictions_0.csv')
pred1_df = pd.read_csv('job_predictions_1.csv')
pred2_df = pd.read_csv('job_predictions_2.csv')
pred3_df = pd.read_csv('job_predictions_3.csv')
pred4_df = pd.read_csv('job_predictions_4.csv')
```

```python
print(pred0_df.head(4))
print(pred1_df.head(4))
print(pred2_df.head(4))
print(pred3_df.head(4))
print(pred4_df.head(4))
```

```python
new_pred_df = pred0_df.copy()
new_pred_df.rename(columns = {'prediction': 'pred0'}, inplace = True)
new_pred_df['pred1'] = pred1_df['prediction']
new_pred_df['pred2'] = pred2_df['prediction']
new_pred_df['pred3'] = pred3_df['prediction']
new_pred_df['pred4'] = pred4_df['prediction']
```

```python
new_pred_df
```

```python
new_pred_df['avg_pred'] = new_pred_df.iloc[:, 1:].mean(axis = 1)
```
