```python
columns_to_join = [
    'LanguageHaveWorkedWith', 'LanguageWantToWorkWith',
    'DatabaseHaveWorkedWith', 'DatabaseWantToWorkWith',
    'PlatformHaveWorkedWith', 'PlatformWantToWorkWith',
    'WebframeHaveWorkedWith', 'WebframeWantToWorkWith',
    'MiscTechHaveWorkedWith', 'MiscTechWantToWorkWith',
    'ToolsTechHaveWorkedWith', 'ToolsTechWantToWorkWith'
]

# Join the information from the specified columns into a single string
# You can adjust the separator as needed, here I'm using ', '
joined_info = user_df[columns_to_join].apply(lambda row: ', '.join(row.dropna()), axis=1)


# If you want to create a new DataFrame with just this joined information
joined_df = joined_info.to_frame(name='JoinedInfo')
# If you want to add this joined information as a new column in the DataFrame
user_df['JoinedInfo'] = joined_info




```

```python
columns_to_join2 = [
   'LanguageWantToWorkWith',
    'DatabaseWantToWorkWith',
    'PlatformWantToWorkWith',
    'WebframeWantToWorkWith',
    'MiscTechWantToWorkWith',
    'ToolsTechWantToWorkWith'
]

# Join the information from the specified columns into a single string
# You can adjust the separator as needed, here I'm using ', '
joined_info2 = user_df[columns_to_join2].apply(lambda row: ', '.join(row.dropna()), axis=1)

# If you want to create a new DataFrame with just this joined information
joined_df2 = joined_info2.to_frame(name='JoinedInfo2')
# If you want to add this joined information as a new column in the DataFrame
user_df['JoinedInfo2'] = joined_info2

```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
user_id = 41504
row_number = user_df[user_df['UserId'] == user_id].index[0]
print(row_number)
```

```python


user_info = joined_df["JoinedInfo"][row_number]
user_info2 = joined_df2["JoinedInfo2"][row_number]
```

```python

user_embedding= model.encode(user_info)
user_embedding2 = model.encode(user_info2)
```

```python
user_embedding.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


user_embedding_2d = user_embedding.reshape(1, -1)

# Convert job_df["job_embeddings"] to a 2D array if it's not already
# This step depends on how job_df["job_embeddings"] is stored
# Function to convert string representation of a list into an actual list of floats
def convert_to_list(s):
    # Remove brackets and split the string by spaces
    elements = s.strip('[]').split()
    # Convert each element to float and return as a list
    return [float(e) for e in elements]

# Apply the conversion function to each element in the 'job_embeddings' column
job_df['job_embeddings'] = job_df['job_embeddings'].apply(convert_to_list)

# Convert the list of embeddings to a 2D NumPy array
job_embeddings_2d = np.vstack(job_df['job_embeddings'])

# Now you can compute the cosine similarity
similarity_scores = cosine_similarity(user_embedding.reshape(1, -1), job_embeddings_2d)[0]

# Add similarity scores to the DataFrame and sort
job_df['Similarity'] = similarity_scores
top_n = 5
similarity = job_df.sort_values(by='Similarity', ascending=False).head(top_n)

# Print the top n similar jobs
print(similarity[['jobdescription', 'Similarity']])
```

```python
top_jobs = job_df.sort_values(by='Similarity', ascending=False).head(top_n)

# Get the bottom 5 similar jobs
bottom_jobs = job_df.sort_values(by='Similarity', ascending=False).tail(top_n)

# Concatenate the top 5 and bottom 5
top_bottom_jobs = pd.concat([top_jobs, bottom_jobs])

# Save the result to a CSV file
top_bottom_jobs.to_csv('/Users/maxxtucker/Documents/GitHub/BT4222/sbert evaluation/jobs_similarity_1.csv', index=False)
```
