Purpose: Encode user ID and job ID into Integers

# Load dataset

```python
import pandas as pd
```

```python
jobEmbedding = pd.read_csv('Embedding/embedded_jobs.csv')
userEmbedding = pd.read_csv('Embedding/embedded_users.csv')
```

```python
print(jobEmbedding.shape)
print(userEmbedding.shape)
```

```python
userEmbedding.head(3)
```

# Encode UserId

```python
# drop ResponseId column and convert to integer values from 1 to len(userEmbedding)
userEmbedding = userEmbedding.drop(columns=['ResponseId'])
userEmbedding = userEmbedding.reset_index()
userEmbedding = userEmbedding.rename(columns={'index': 'UserId'})
```

```python
userEmbedding.head(3)
```

```python
# Save the embedded users to a csv file
userEmbedding.to_csv('Embedding/embedded_users_IdEncoded.csv', index=False)
```

# Encode JobId

```python
jobEmbedding.head(3)
```

```python
# drop jobid column and convert to integer values from 1 to len(jobEmbedding)
jobEmbedding = jobEmbedding.drop(columns=['uniq_id'])
jobEmbedding = jobEmbedding.drop(columns=['jobid'])
jobEmbedding = jobEmbedding.reset_index()
jobEmbedding = jobEmbedding.rename(columns={'index': 'JobId'})
```

```python
jobEmbedding.head(3)
```

```python
jobEmbedding.to_csv('Embedding/embedded_jobs_IdEncoded.csv', index=False)
```
